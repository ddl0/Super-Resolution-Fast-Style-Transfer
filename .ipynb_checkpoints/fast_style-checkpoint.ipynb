## Reflection Padding

```python
class ReflectionPadding2D(Layer):
    def __init__(self, padding=(1, 1), **kwargs):
        self.padding = tuple(padding)
        self.input_spec = [InputSpec(ndim=4)]
        super(ReflectionPadding2D, self).__init__(**kwargs)

    def get_output_shape_for(self, s):
        return (s[0],
                s[1] + 2 * self.padding[0],
                s[2] + 2 * self.padding[1],
                s[3])

    def call(self, x, mask=None):
        w_pad, h_pad = self.padding
        return tf.pad(x, [[0, 0], [h_pad, h_pad], [w_pad, w_pad], [0, 0]], mode='REFLECT')
```

## Style loss

Compute the style loss. It's the difference between the gram matrixes of all style layers of the style target and the current image.

```python
style_layer_names = ['block1_conv2',
                     'block2_conv2', 'block4_conv3', 'block5_conv3']
style_layers = [vgg.get_layer(name).output for name in style_layer_names]
style_layers
```

```python
style_extract_model = Model(vgg.input, style_layers)
styles_precomputed = style_extract_model.predict_on_batch(
    style_image[np.newaxis])
[np.squeeze(style).shape for style in styles_precomputed]
styles_precomputed = [tf.Variable(layer) for layer in styles_precomputed]
```

```python
def gram_matrix(layer):
    ''' Compute the gram activations of a layer.
    Args:
        layer: Tensor with shape (batch_size, width, height, channels)
    '''
    # Swap axes to (batch, channels, width, height)
    layer = tf.transpose(layer, perm=(0, 3, 1, 2))
    shape = layer.get_shape()

    # Combine the width and height dimensions
    layer = tf.reshape(layer,
                       shape=tf.stack([-1, shape[1], shape[2] * shape[3]]))

    shapes_product = tf.cast(
        shape[1] * shape[2] * shape[3], dtype=tf.float32)

    transposed_layer = tf.transpose(layer, perm=(0, 2, 1))
    gram = tf.matmul(layer, transposed_layer) / shapes_product
    return gram
```

```python
def mean_square_batch(x):
    dims = list(range(1, len(x.get_shape())))
    means = tf.reduce_mean(x ** 2, dims)
    sqrt = tf.sqrt(means)
    expanded = tf.expand_dims(sqrt, axis=0)
    return expanded


def style_loss_layer(layer1, layer2):
    return mean_square_batch(layer1 - layer2)
```

```python
def style_loss(image):
    loss = 0
    style_tensors = style_extract_model(image)
    for layer1, layer2 in zip(styles_precomputed, style_tensors):
        layer_loss = style_loss_layer(
            gram_matrix(layer1),
            gram_matrix(layer2))
        loss += layer_loss
    return loss
```

## Content loss

```python
content_layer = vgg.get_layer('block2_conv2').output
content_model = Model(vgg_inp, content_layer)
content_model.summary()
```

```python
def content_loss(image1, image2):
    image1_activations = content_model(image1)
    image2_activations = content_model(image2)
    return mean_square_batch(image1_activations - image2_activations)
#     return K.sqrt(K.mean(K.square(image1_activations - image2_activations),
#                          axis=(1, 2)))
```

## Total variance loss

```python
def total_variation_loss(image, image_shape=None):
    if image_shape:
        width, height, channels = image_shape
    else:
        # Try to get the image size from the tensor
        dims = image.get_shape()
        width = dims[1].value
        height = dims[2].value
        channels = dims[3].value

    tv_x_size = width * (height - 1) * channels
    tv_y_size = (width - 1) * height * channels

    return tf.expand_dims(
        (tf.reduce_sum(tf.abs(image[:, 1:, :, :] - image[:, :-1, :, :])) +
         tf.reduce_sum(tf.abs(image[:, :, 1:, :] - image[:, :, :-1, :]))),
        axis=0)
```

## Fast Style Network

```python
def conv_block(inp, filters, size,
               stride=1, activation=True, border_mode='same'):
    ''' Define a convolutional block for the upscale network.
    '''
    x = Convolution2D(nb_filter=filters, nb_row=size[0], nb_col=size[1],
                      border_mode=border_mode, subsample=(stride, stride))(inp)
    x = BatchNormalization()(x)
    if activation:
        return Activation(activation='relu')(x)
    else:
        return x
```

```python
def res_block_crop(inp, filters):
    x = conv_block(inp, filters=filters, size=(3, 3), border_mode='valid')
    x = conv_block(x, filters=filters, size=(3, 3), border_mode='valid',
                   activation=False)
    # since the convolutions will cut off 2 pixels each
    inp = Lambda(lambda x: x[:, 2:-2, 2:-2])(inp)
    x = merge([x, inp], mode='sum')
    return x
```

```python
def deconv_block(inp, filters, size, output_shape, stride=2):
    x = Deconvolution2D(nb_filter=filters,
                        nb_row=size[0], nb_col=size[1],
                        output_shape=output_shape,
                        subsample=(stride, stride), border_mode='same')(inp)
    x = BatchNormalization()(x)
    return Activation('relu')(x)
```

```python
def up_block(inp, filters, size):
    x = UpSampling2D(size=(2,2))(inp)
    x = Convolution2D(filters, size[0], size[1], border_mode='same')(x)
    x = BatchNormalization()(x)
    return Activation('relu')(x)
```

```python
x = ReflectionPadding2D(padding=(40, 40))(inp)
x = conv_block(x, filters=64, size=(9, 9))
x = conv_block(x, filters=64, size=(3, 3), stride=2)
x = conv_block(x, filters=64, size=(3, 3), stride=2)
for i in range(5):
    x = res_block_crop(x, 64)
x = up_block(x, 64, (3, 3))
x = up_block(x, 64, (3, 3))
# x = deconv_block(x, 64, size=(3, 3), output_shape=(None, 144, 144, 64))
# x = deconv_block(x, 64, size=(3, 3), output_shape=(None, 288, 288, 64))
x = Convolution2D(nb_filter=3, nb_row=9, nb_col=9,
                  activation='tanh', border_mode='same')(x)

print(x)
def deprocess(x):
    x = (x + 1) * 127.5
#     x.set_shape((None,) + image_size + (3,))
    return x


fast_output = Lambda(deprocess)(x)

fast_model = Model(inp, fast_output)
```

## Full Network

```python
def total_loss(args):
    inp = args[0]
    output = args[1]
#     output = fast_model(inp)
    style_loss_tensor = style_loss(output)
    content_loss_tensor = content_loss(inp, output)
#     total_variance_loss_tensor = total_variation_loss(output)
    content_weight = 1e0
    style_weight = 1e-3
    tv_weight = 0  # 1e-2
    loss = (style_weight * style_loss_tensor
            + content_weight * content_loss_tensor)
#             + tv_weight * total_variance_loss_tensor)
    return loss
```

```python
loss_lambda = Lambda(total_loss)([inp, fast_output])
final_model = Model([inp, vgg_inp], loss_lambda)
```

```python
final_model.save_weights('reset_weights.hdf5')
```

## Fit

```python
final_model.load_weights('reset_weights.hdf5')
lr = 1e-3
final_model.compile(optimizer=Adam(lr=lr), loss='mae')
```

```python
targets = np.zeros((content_images.shape[0], 1))
params = {'verbose': 0, 'callbacks': [TQDMNotebookCallback(leave_inner=True)]}
final_model.fit([content_images, content_images], targets,
                batch_size=4, nb_epoch=1, **params)
```
